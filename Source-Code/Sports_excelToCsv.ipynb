```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO

# Get uploaded file name
file_name = list(uploaded.keys())[0]

# Read Excel
df = pd.read_excel(file_name)

print("🔍 Original Dataset Preview")
display(df.head())

print("\n📌 Dataset Info:")
df.info()

print("\n📊 Missing Values:")
display(df.isnull().sum())

print("\n📈 Statistical Summary:")
display(df.describe())
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

if len(numeric_cols) > 0:
    col = numeric_cols[0]  # first numeric column

    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col} (Original Data)")
    plt.show()
```

```python
csv_file = "converted_file.csv"

# Convert
df.to_csv(csv_file, index=False, encoding='utf-8')

print("✅ Conversion Done!")
```

```python
df_csv = pd.read_csv(csv_file)

print("🔍 CSV Dataset Preview")
display(df_csv.head())

print("\n📌 CSV Info:")
df_csv.info()

print("\n📊 Missing Values (CSV):")
display(df_csv.isnull().sum())

print("\n📈 Statistical Summary (CSV):")
display(df_csv.describe())
```

```python
numeric_cols_csv = df_csv.select_dtypes(include=['int64', 'float64']).columns

if len(numeric_cols_csv) > 0:
    col2 = numeric_cols_csv[0]

    plt.figure()
    sns.histplot(df_csv[col2], kde=True)
    plt.title(f"Distribution of {col2} (CSV Data)")
    plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = df.corr(numeric_only=True)

if not correlation_matrix.empty:
    plt.figure()
    sns.heatmap(correlation_matrix, annot=True)
    plt.title("Correlation Heatmap (Original Data)")
    plt.show()
else:
    print("No numeric columns found in the DataFrame to generate a correlation heatmap.")
```
